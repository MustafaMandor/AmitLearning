```python
import json
import numpy as np
```

```python
with open('data.json') as file:
    data=json.load(file)
```

```python
data
```

```python
training_examples=[]
training_labels=[]
for example in data['data']:
    for pattern in example['patterns']:
        training_examples.append(pattern)
        training_labels.append(example['label'])
```

```python
training_examples
```

```python
training_labels
```

```python
def clean_text(sent):
    result=[]
    for example in sent:
        result.append(example.lower())
    return result
```

```python
training_examples=clean_text(training_examples)
```

```python
len(set(training_labels))
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
training_labels=le.fit_transform(training_labels)
training_labels=to_categorical(training_labels)
training_labels
```

```python
import pickle
pickle.dump(le,open("training_labels_le.pkl","wb"))
```

```python
from tensorflow.keras.preprocessing.text import one_hot
sequence=[one_hot(example,100,filters='') for example in training_examples]
sequence
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(num_words=100)
tokenizer.fit_on_texts(training_examples)
sequences=tokenizer.texts_to_sequences(training_examples)
print(sequences)
```

```python
pickle.dump(tokenizer,open("tokenizer.pkl","wb"))
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_sequence=pad_sequences(sequences,maxlen=20)
print(padded_sequence)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Embedding,Dense,Flatten
model= Sequential()
model.add(Embedding(100,10,input_length=20))
model.add(Flatten())
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(8,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(padded_sequence,training_labels,epochs=400)
```

```python
pickle.dump(model,open('model.pkl','wb'))
```

```python
inp='bye'
in_tok=tokenizer.texts_to_sequences([inp])
sent=pad_sequences(in_tok,maxlen=20)
```

```python
sent
```

```python
model.predict(sent)
```

```python
y_pred=np.argmax(model.predict(sent))
y_pred
```

```python
output=le.inverse_transform(np.array(y_pred).reshape(1))
output
```

```python
def user_chat():
    while True:
        print('User:',end='')
        inp=input()
        if inp.lower()=='quit':
            break
        in_tok=tokenizer.texts_to_sequences([inp])
        sent=pad_sequences(in_tok,maxlen=20)
        y_pred=np.argmax(model.predict(sent))
        output=le.inverse_transform(np.array(y_pred).reshape(1))
        
        for label in data['data']:
            if label['label']==output:
                print("Chat bot:",np.random.choice(label['responses']))


        
```

```python
# Drop Nan values
data.dropna(inplace=True)
```

```python
## Two same functions to separate the dates and add each number(days,month,year) in a list and make a column with each list(days,months,years)
def date_split_1(date_1):
    days=[]
    months=[]
    years=[]
    for d in date_1:
        day,month,year=d.split('-')
        days.append(day)
        months.append(month)
        years.append(year)
    data['Day_of_Surgery']=days
    data['Month_of_Surgery']=months
    data['Year_of_Surgery']=years
def date_split_2(date_2):
    days=[]
    months=[]
    years=[]
    for d in date_2:
            day,month,year=d.split('-')
            days.append(day)
            months.append(month)
            years.append(year)
    data['Day_of_Last_Visit']=days
    data['Month_of_Last_Visit']=months
    data['Year_of_Last_Visit']=years
```

```python
date_split_1(data['Date_of_Surgery'].tolist())
date_split_2(data['Date_of_Last_Visit'].tolist())
```

```python
data
```

```python
#Drop the old dates
data.drop(['Date_of_Surgery','Date_of_Last_Visit'],axis=1,inplace=True)
```

```python
data.info()
```

```python
#Making label encoder
from sklearn.preprocessing import LabelEncoder
Le=LabelEncoder()
cols=['Gender','Tumour_Stage','Histology','ER status','PR status','HER2 status','Surgery_type','Patient_Status']
for col in cols:
    data[col]=Le.fit_transform(data[col])
```
